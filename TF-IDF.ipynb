```python
from collections import Counter         #importing all the required lib
from scipy.sparse import csr_matrix
import math
import operator
from sklearn.preprocessing import normalize
import numpy
import warnings
warnings.filterwarnings("ignore")
```

```python
corpus = [
     'this is the first document',
     'this document is the second document',
     'and this is the third one',
     'is this the first document',
]
```

```python
def IDF(corpus,unique_words):  #Function for caluclation the IDF score
    idf = {}                   #I will store the words and its score in this dict
    x = len(corpus)
    for i in unique_words:
        count = 0
        for j in corpus:
            if i in j.split():
                count = count+1
            idf[i]=(math.log ((1+x)/(count+1)))+1
    return idf
```

```python
def fit(dataset):    
    unique_words = set() # at first we will initialize an empty set
    # check if its list type or not
    if isinstance(dataset, (list,)):
        for row in dataset: # for each review in the dataset
            for word in row.split(" "): # for each word in the review. #split method converts a string into list of words
                if len(word) < 2:
                    continue
                unique_words.add(word)
        unique_words = sorted(list(unique_words))
        vocab = {j:i for i,j in enumerate(unique_words)}
        idf_values = IDF(corpus,unique_words)      #calling the IDF function to calulate Idf score for each unique word
        return vocab, idf_values
    else:
        print("you need to pass list of sentance")
```

```python
 Vocabulary, idf_of_vocabulary=fit(corpus)   #callind the fit function
```

```python
 print(list(Vocabulary.keys())) 
```

```python
 print(list(idf_of_vocabulary.values()))
```

```python
def transform(corpus,vocabulary,idf_values):   #the Transform function
    sparse_matrix = csr_matrix( (len(corpus), len(vocabulary)) , dtype = numpy.float) #compressed sparse row matrix
    
    for row in range(0,len(corpus)): 
        num_of_word = Counter(corpus[row].split())
        for word in corpus[row].split():
            if word in list (vocabulary.keys()):
                
                tf_idf_value = (num_of_word[word]/len(corpus[row].split()))*(idf_values[word]) #calculating TF-Idf Value
                
                sparse_matrix[row,vocabulary[word]] = tf_idf_value
    
    print("NORM\n", normalize(sparse_matrix, norm ="l2", axis=1 , copy = True, return_norm = False)) #using l2 norm
    result = normalize(sparse_matrix, norm ="l2", axis=1 , copy = True, return_norm = False)
    return result
```

```python
result  = transform(corpus,Vocabulary,idf_of_vocabulary)
print(result)
```

```python
 print(result[0].toarray())
```
